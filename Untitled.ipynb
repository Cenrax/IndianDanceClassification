```python
test_dir = 'test/'
test_fnames = os.listdir(test_dir)
```

```python
img_width = 224
img_height = 224
```

```python
def train_data_prep(list_of_images, train, train_dir):
    train_data=[]
    train_label=[]
    for image in list_of_images:
        train_data.append(cv2.resize(cv2.imread(train_dir+image), (img_width,img_height),interpolation=cv2.INTER_CUBIC))
        if image in list(train['Image']):
            train_label.append(train.loc[train['Image'] == image, 'target'].values[0])
            
    return train_data, train_label
```

```python
def test_data_prep(list_of_images, test_dir):
    test_data = []
    for image in list_of_images:
        test_data.append(cv2.resize(cv2.imread(test_dir+image), (img_width,img_height),interpolation=cv2.INTER_CUBIC))
    
    return test_data
    
```

```python
testing_data = test_data_prep(test_fnames,test_dir)
```

```python
training_data, training_labels = train_data_prep(train_fnames,train, train_dir)
```

```python
training_labels
```

```python
mapping = {'bharatanatyam':0,'kathak':1,'kathakali':2,'kuchipudi':3,'manipuri':4,'mohiniyattam':5,'odissi':6,'sattriya':7}
```

```python
def show_batch(image_batch, label_batch):
    plt.figure(figsize=(12,12))
    for n in range(25):
        
        ax = plt.subplot(5,5,n+1)
        plt.imshow(image_batch[n])
        plt.title(label_batch[n].title())
        plt.axis('off')
```

```python
show_batch(training_data, training_labels)
```

```python
le = LabelEncoder()
training_labels = le.fit_transform(training_labels)
```

```python
training_labels
```

```python
X_train, X_val, y_train, y_val = train_test_split(training_data, training_labels, test_size=0.33, random_state=42)
```

```python
train_datagenerator = ImageDataGenerator(
                    rescale=1./255,
featurewise_center=False,
samplewise_center=False,
featurewise_std_normalization=False,
samplewise_std_normalization=False,
rotation_range=50,
zoom_range=0.2,
width_shift_range=0.10,
height_shift_range=0.10,
horizontal_flip=True,
vertical_flip=False)
```

```python
val_datagenerator = ImageDataGenerator(rescale=1. / 255)
```

```python
train_datagenerator.fit(X_train)
val_datagenerator.fit(X_val)
X_train = np.array(X_train)
X_val = np.array(X_val)
```

```python
X_val.shape
```

```python
vggmodel = VGG16(weights='imagenet', include_top=False, input_shape = (224,224,3), pooling = 'max')
```

```python
vggmodel.summary()
```

```python
vggmodel.trainable = False
model = Sequential([
    vggmodel,
    Dense(1024, activation='relu'),
    Dropout(0.2),
    Dense(256,activation='relu'),
    Dropout(0.2),
    Dense(8,activation='softmax'),
])
```

```python
reduce_learning_rate = ReduceLROnPlateau(monitor = 'loss',
                                        factor = 0.1,
                                        patience=2,
                                        cooldown=2,
                                        min_lr=0.00001,
                                        verbose=1)

callbacks = [reduce_learning_rate]
```

```python
model.compile(optimizer = 'adam', loss='categorical_crossentropy', metrics = ['accuracy'])
history = model.fit_generator(
train_datagenerator.flow(X_train, to_categorical(y_train,8), batch_size=16),
validation_data=val_datagenerator.flow(X_val, to_categorical(y_val,8), batch_size=16),
verbose=2,
epochs=30,
callbacks=callbacks)
```

```python
import matplotlib.image as mpimg
```

```python
import matplotlib.pyplot as plt

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(len(acc))
```

```python
plt.plot(epochs,acc)
plt.plot(epochs,val_acc)
plt.title('Accuarcy')
plt.figure()
```
